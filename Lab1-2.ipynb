Ensure your results match for both methods.

```python
# preprocessing for both
import pandas as pd
import numpy as np

## loading the world population data
filenameWorld = "world_population.csv"
worldDf = pd.read_csv(filenameWorld)   ## pandas data frame for world population data
worldData = np.array(worldDf)          ## numPy array with the  world population data
worldDl = [list(x) for x in worldData]  ## List of lists with the world population data

## loading the corruption index  data
filenameCorruption = "corruption_data.csv"
corrDf = pd.read_csv(filenameCorruption)   ## pandas data frame for corruption index data
corrDf['country'] = [x.strip() for x in corrDf["region_name"]]
corrData = np.array(corrDf)              ## numPy array with the  corruption index data
corrDl = [list(x) for x in corrData]     ## List of lists with the corruption index data

# clean
world_clean = worldDf.copy()
world_clean["Country"] = world_clean["Country"].astype(str).str.strip()
world_clean["Continent"] = world_clean["Continent"].astype(str).str.strip()

corr_clean = corrDf.copy()
corr_clean["country"] = corr_clean["country"].astype(str).str.strip()
corr_clean["country"] = corr_clean["country"].replace({
    "Democratic Republic of the Congo" : "DR Congo",
    "São Tomé and Príncipe" : "Sao Tome and Principe"
})
print("")
```

```python
### Your code goes here: Solution 1
africa_no_corr_loop = []

african_countries = world_clean.loc[
    world_clean["Continent"] == "Africa", "Country"
].tolist()

corr_countries = corr_clean["country"].tolist()

for c in african_countries:
    found = False
    for cc in corr_countries:
        if c == cc:
            found = True
            break
    if not found:
        africa_no_corr_loop.append(c)

africa_no_corr_loop = sorted(set(africa_no_corr_loop))
```

```python
### Your code goes here: Solution 2
africa_no_corr_lib = (
    world_clean.loc[world_clean["Continent"] == "Africa", "Country"] # list o' country name
    .loc[~world_clean.loc[world_clean["Continent"] == "Africa", "Country"] # keep only rows where False 
         .isin(corr_clean["country"])]
    .dropna()
    .sort_values()
    .unique()
    .tolist()
)
```

```python
# output for both
print("African countries with no corruption data (Loop approach):")
for c in africa_no_corr_loop:
    print(c)

print("\nAfrican countries with no corruption data (Library approach):")
for c in africa_no_corr_lib:
    print(c)
```

**Question 9:** For every continent, identify the least corrupt country (or countries) in 2020 (include all ties).

**Output Requirements**: For each matching country, report:

1) Continent

2) Country Name

3) 2020 Corruption Index

4) 2020 Population

**Sorting (Mandatory)**:

1) First, sort by Continent in alphabetical order (A–Z).

2) Second, for any ties within a continent, sort by 2020 Population in descending order (largest to smallest).

**Restriction (Pandas API)**: You must use the pandas (and/or NumPy) API to solve this. Avoid explicit Python for loops.

**Extra Credit**: Try to accomplish this in a single statement using method chaining (e.g., df.groupby(...).filter(...)). While splitting it into multiple steps is allowed, the single-statement approach more desirable.

```python
import pandas as pd
import numpy as np

## loading the world population data
filenameWorld = "world_population.csv"
worldDf = pd.read_csv(filenameWorld)   ## pandas data frame for world population data
worldData = np.array(worldDf)          ## numPy array with the  world population data
worldDl = [list(x) for x in worldData]  ## List of lists with the world population data

## loading the corruption index  data
filenameCorruption = "corruption_data.csv"
corrDf = pd.read_csv(filenameCorruption)   ## pandas data frame for corruption index data
corrDf['country'] = [x.strip() for x in corrDf["region_name"]]
corrData = np.array(corrDf)              ## numPy array with the  corruption index data
corrDl = [list(x) for x in corrData]     ## List of lists with the corruption index data

# code start
result = (
    worldDf.assign(
        Country=lambda d: d["Country"].astype(str).str.strip(),
        Continent=lambda d: d["Continent"].astype(str).str.strip(),
        Pop2020=lambda d: pd.to_numeric(d["2020 Population"], errors="coerce"),
    )
    .merge(
        corrDf.assign(country=lambda d: d["country"].astype(str).str.strip()),
        left_on="Country",
        right_on="country",
        how="inner",
    )
    .assign(CPI2020=lambda d: pd.to_numeric(d["2020"], errors="coerce"))
    .dropna(subset=["CPI2020", "Pop2020"])
    .loc[
        lambda d: d["CPI2020"].eq(d.groupby("Continent")["CPI2020"].transform("max")),
        ["Continent", "Country", "CPI2020", "Pop2020"],
    ]
    .sort_values(["Continent", "Pop2020"], ascending=[True, False])
    .rename(
        columns={
            "CPI2020": "2020 Corruption Index",
            "Pop2020": "2020 Population",
        }
    )
    .reset_index(drop=True)
)

result
```

**Question 10:** For every year documented in the corruption dataset (2012–2021), identify the country with the smallest land area that maintained a "Good" Corruption Index ($\ge 70$) during that specific year.

 **Objective**: You are producing a year-by-year report. For each year, you must filter for "Good" countries, then find the one among them with the minimum area.
 
 **Output Requirements**: For each identified country, report:
 1) The Year
 2) Country Name
 3) Area (from the population/metadata file)
 4) Average Corruption Index: Calculate the mean score for that specific country across all years available in the dataset (2012–2021).
 
 **Technical Constraints**: 
 1) You must link the area data (Population file) with the annual scores (Corruption file).
 2) Aggregation: Note that the "Average Index" is a global average for that country's history in the file, not just the score for that specific year.
 3) Handling Ties: If two "Good" countries have the exact same smallest area in a given year, report both.
 
 **Sorting**: Sort the final output chronologically by Year.

```python
import pandas as pd
import numpy as np

## loading the world population data
filenameWorld = "world_population.csv"
worldDf = pd.read_csv(filenameWorld)   ## pandas data frame for world population data
worldData = np.array(worldDf)          ## numPy array with the  world population data
worldDl = [list(x) for x in worldData]  ## List of lists with the world population data

## loading the corruption index  data
filenameCorruption = "corruption_data.csv"
corrDf = pd.read_csv(filenameCorruption)   ## pandas data frame for corruption index data
corrDf['country'] = [x.strip() for x in corrDf["region_name"]]
corrData = np.array(corrDf)              ## numPy array with the  corruption index data
corrDl = [list(x) for x in corrData]     ## List of lists with the corruption index data

# clean world
world_clean = (
    worldDf.assign(
        Country=lambda d: d["Country"].astype(str).str.strip(),
        Area=lambda d: pd.to_numeric(d["Area (km²)"], errors="coerce"),
    )
    .loc[:, ["Country", "Area"]]
)

# clean corr
corr_long = (
    corrDf.assign(country=lambda d: d["country"].astype(str).str.strip())
    .melt(
        id_vars=["country"],
        value_vars=[str(y) for y in range(2012, 2022)],
        var_name="Year",
        value_name="CPI",
    )
    .assign(
        Year=lambda d: d["Year"].astype(int),
        CPI=lambda d: pd.to_numeric(d["CPI"], errors="coerce"),
    )
)

# compute avg
avg_cpi = (
    corr_long.groupby("country", as_index=False)["CPI"]
    .mean()
    .rename(columns={"CPI": "Average Corruption Index"})
)

# merge
merged = (
    corr_long.merge(world_clean, left_on="country", right_on="Country", how="inner")
    .merge(avg_cpi, on="country", how="inner")
)

# keep good countries
good = merged[merged["CPI"] >= 70]

# find min area
result = (
    good.loc[
        good["Area"].eq(good.groupby("Year")["Area"].transform("min")),
        ["Year", "country", "Area", "Average Corruption Index"],
    ]
    .rename(columns={"country": "Country Name"})
    .sort_values("Year")
    .reset_index(drop=True)
)

result
```
