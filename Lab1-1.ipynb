**Question 7:** Find the percent of male and female students in each school that consume considerable quantities of alcohol (value of 3 or above) _both_ on weekends and weekdays.
(your output shall consist of four rows of data).

```python
### Your code goes here
q7_df = df.copy()

female_df = q7_df[q7_df["sex"] == "F"]
male_df = q7_df[q7_df["sex"] == "M"]

female_gp_df = female_df[female_df["school"] == "GP"]
female_ms_df = female_df[female_df["school"] == "MS"]

male_gp_df = male_df[male_df["school"] == "GP"]
male_ms_df = male_df[male_df["school"] == "MS"]

female_GP_dalc = round((female_gp_df[female_gp_df["Dalc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
female_MS_dalc = round((female_ms_df[female_ms_df["Dalc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
female_GP_walc = round((female_gp_df[female_gp_df["Walc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
female_MS_walc = round((female_ms_df[female_ms_df["Walc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)

male_GP_dalc = round((male_gp_df[male_gp_df["Dalc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
male_MS_dalc = round((male_ms_df[male_ms_df["Dalc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
male_GP_walc = round((male_gp_df[male_gp_df["Walc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)
male_MS_walc = round((male_ms_df[male_ms_df["Walc"] > 2].shape[0] / q7_df.shape[0]) * 100, 2)


result_q7 = pd.DataFrame({
    "sex" : ["Male", "Male", "Female", "Female"],
    "school" : ["GP", "MS", "GP", "MS"],
    "Dalc %" : [male_GP_dalc, male_MS_dalc, female_GP_dalc, female_MS_dalc],
    "Walc %" : [male_GP_walc, male_MS_walc, female_GP_walc, female_MS_walc]
})

result_q7
```
